## Comments

In their seminal book, *Structure and Interpretation of Computer
Programs*, Abelson and Sussman claim that,

> Programs must be written for people to read, and only incidentally for
> machines to execute.

It is often the case that programs need explanation beyond the code you
have typed out in order for others (or even yourself) to understand
them. This is what *commenting* is for.

``` python
# This is a comment
x is ask "What is x?" # This is also a comment
```

Comments are text that is ignored by the program so you can write
whatever you need. Sometimes it can even be useful to draw diagrams in
comments!

``` python
#       _
#      | |
#      | | __ _ _ __ ___   ___  ___
#  _   | |/ _` | '_ ` _ \ / _ \/ __|
# | |__| | (_| | | | | | |  __/\__ \
#  \____/ \__,_|_| |_| |_|\___||___/
```
